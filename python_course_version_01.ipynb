```python
import numpy as np

# Create a 3x3 matrix of random integers between 1 and 10
matrix = np.random.randint(1, 11, size=(3, 3))
print("Original matrix:")
print(matrix)

# Compute and print transpose
transpose = matrix.T
print("\nTranspose:")
print(transpose)

# Compute and print determinant
determinant = np.linalg.det(matrix)
print(f"\nDeterminant: {determinant}")
```

### 5. Practical Application

Let's put it all together with a practical example. We'll create a simple neural network using TensorFlow, train it on some random data, and visualize the results.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
X = np.linspace(-1, 1, 100).reshape(-1, 1)
y = 2 * X + np.random.randn(*X.shape) * 0.3

# Create a simple model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,))
])

# Compile the model
model.compile(optimizer='sgd', loss='mse')

# Train the model
history = model.fit(X, y, epochs=50, verbose=0)

# Plot the results
plt.scatter(X, y, label='Data')
plt.plot(X, model.predict(X), color='red', label='Prediction')
plt.legend()
plt.title('Simple Linear Regression')
plt.show()

# Plot the loss over epochs
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

__Final Exercise:__ Modify the above example to create a slightly more complex model with one hidden layer of 10 neurons and a ReLU activation function. Train it on the same data and compare the results with the simple linear model.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
X = np.linspace(-1, 1, 100).reshape(-1, 1)
y = 2 * X + np.random.randn(*X.shape) * 0.3

# Create a simple model (same as before)
simple_model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,))
])

# Compile and train the simple model
simple_model.compile(optimizer='sgd', loss='mse')
simple_history = simple_model.fit(X, y, epochs=50, verbose=0)

# Create a more complex model
complex_model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

# Compile and train the complex model
complex_model.compile(optimizer='adam', loss='mse')
complex_history = complex_model.fit(X, y, epochs=50, verbose=0)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Data')
plt.plot(X, simple_model.predict(X), color='red', label='Simple Model')
plt.plot(X, complex_model.predict(X), color='green', label='Complex Model')
plt.legend()
plt.title('Comparison of Simple and Complex Models')
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Plot the loss over epochs for both models
plt.figure(figsize=(10, 6))
plt.plot(simple_history.history['loss'], label='Simple Model Loss')
plt.plot(complex_history.history['loss'], label='Complex Model Loss')
plt.title('Model Loss Comparison')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# Print final loss for both models
print(f"Final loss for simple model: {simple_history.history['loss'][-1]:.4f}")
print(f"Final loss for complex model: {complex_history.history['loss'][-1]:.4f}")
```
